# Ejercicio en Clase

En este ejercicio van a analizar que ocurre con la ecuacion normal de la regresion lineal cuando una variable de entrada escala progresivamente en multiplos de 10. La meta es observar en que punto el calculo comienza a presentar problemas numericos, que hacen que se vuelva inestable o incalculable!

Intrucciones:
1. Crear una lista de tamaño j, con valores 10^j, hasta un j = 20
2. Iterar esta lista, en cada iteracion: crear la matriz X con el tamano de la iteracion, y hacer el calculo de theta_mejor
3. En cada iteracion calcular el tiempo transcurrido desde el inicio hasta el final de la iteracion (calculo), imprimir el theta y el tiempo transcurrido, imprimir en que iteracion estan!
4. Al final reportar/imprimir cuando la computadora/kernel de python deja de responder, M maximo iterado...

```python
from sklearn.preprocessing import add_dummy_feature
import time
import numpy as np
max_j_logrado = 0
#Maximo rango de iteracion 7 para 15 gb de ram
for j in range(7):
    m_j = 10 ** j
    print(f"--- Iteracion j={j}, m={m_j} ---")

    try:
        start = time.perf_counter()

        X_j = 7 * np.random.rand(m_j, 1)
        y_j = 10 + 2 * X_j + np.random.randn(m_j, 1)
        X_b_j = add_dummy_feature(X_j)

        theta_j = np.linalg.inv(X_b_j.T @ X_b_j) @ X_b_j.T @ y_j

        end = time.perf_counter()
        elapsed = end - start

        print(f"  theta = {theta_j.ravel()}")
        print(f"  Tiempo: {elapsed:.4f} s")

        max_j_logrado = j

    except Exception as e:
        print(f"  ERROR en j={j}: {e}")
        break

print(f"\nM maximo iterado exitosamente: j={max_j_logrado}, m=10^{max_j_logrado} = {10**max_j_logrado}")
```
